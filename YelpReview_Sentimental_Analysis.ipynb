```python
#Beautiful soup is used to extract data from HTML and XML files
from bs4 import BeautifulSoup 
```

```python
import requests
```

```python
#Getting the data from one of the review website from yelp
r = requests.get('https://www.yelp.com/not_recommended_reviews/vision-auto-glass-burnaby')
```

```python
r.status_code
```

```python
r.text
```

```python
soup = BeautifulSoup(r.text ,'html.parser')
```

```python
#After inspection of the website and figuring out the elements are saved in the review-content division, we are fetching the data from it
soup.findAll(class_ = 'review-content') 
```

```python
type(soup.findAll(class_ = 'review-content'))
```

```python
divs = soup.findAll(class_ = 'review-content')
```

```python
#Appending the information into the list format for usability 
review = []
for div in divs:
    print(div.find('p').text,'\n')
```

```python
for div in divs:
    review.append(div.find('p').text)
```

```python
#Lemmatization process:
#Importing textblob
#pip install textblob
```

```python
from textblob import Word
```

```python
#nltk.download('wordnet')
```

```python
df['clean_reviews'].apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
```

```python
df['lemmatized_reviews'] =df['clean_reviews'].apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
```

```python
df.head()
```

```python
#Review sentimental analysis
from textblob import TextBlob
```

```python
#We get polarity and subjectivity from textblob
df['lemmatized_reviews'].apply(lambda x: TextBlob(x).sentiment)
```

```python
df['polarity'] = df['lemmatized_reviews'].apply(lambda x: TextBlob(x).sentiment[0])
```

```python
df['subjectivity'] = df['lemmatized_reviews'].apply(lambda x: TextBlob(x).sentiment[1])
```
